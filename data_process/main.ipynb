```python
import csv

# Specify CSV file path
csv_file_path = "./nodes.csv"

# Read CSV file
data_nodes = []
with open(csv_file_path, mode='r', encoding='utf-8', newline='') as file:
    reader = csv.DictReader(file)
    for row in reader:
        data_nodes.append({
            "id": int(row["id"]),
            "label": row["label"],
            "group": row["group"]
        })

# Generate JavaScript object
js_object = "var nodes = new vis.DataSet([\n"
for node in data_nodes:
    js_object += f'  {{id: {node["id"]}, label: "{node["label"]}", group: "{node["group"]}"}},\n'
js_object = js_object.rstrip(",\n")  # Remove trailing comma and newline
js_object += "\n]);"

print(js_object)


# Specify JavaScript file path
js_file_path = "./nodes.js"

# Write js_object to JavaScript file
with open(js_file_path, mode='w', encoding='utf-8') as js_file:
    js_file.write(js_object)

print(f"JavaScript object saved to {js_file_path}")
```

```python
import csv

# Specify CSV file path
csv_file_path = "./edges.csv"

# Read CSV file
data_nodes = []
with open(csv_file_path, mode='r', encoding='utf-8', newline='') as file:
    reader = csv.DictReader(file)
    for row in reader:
        data_nodes.append({
            "from": int(row["from"]),
            "to": int(row["to"]),
            "label": row["label"],
            "arrowhead": row["arrowhead"]
        })

# Generate JavaScript object
js_object = "var edges = new vis.DataSet([\n"
for node in data_nodes:
    js_object += f'  {{from: {node["from"]}, to:{node["to"]},label: "{node["label"]}", arrowhead: "{node["arrowhead"]}"}},\n'
js_object = js_object.rstrip(",\n")  # Remove trailing comma and newline
js_object += "\n]);"

print(js_object)


# Specify JavaScript file path
js_file_path = "./edges.js"

# Write js_object to JavaScript file
with open(js_file_path, mode='w', encoding='utf-8') as js_file:
    js_file.write(js_object)

print(f"JavaScript object saved to {js_file_path}")
```
